## Rules for Bad Visualizations

### Rule 1: Show as little data as possible

```python
sales_data = pd.DataFrame([['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'],
                           [400, 200, 600, 800, 700, 1000, 1200, 1050, 900, 600, 300, 500]]).transpose()

sales_data.columns = ['month', 'num_sales']
sales_data.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
ax1 = sales_data.plot(legend = False) 
ax1.set_xlabel('Month')
ax1.set_ylabel("Number of Sales")
plt.xticks([x for x in range(0,12)], sales_data.index, rotation=90)
ax1.set_ylim(0,5000)
plt.show()
```

By changing the limits on the plot, we can lie about how much of an increase or decrease was seen in time periods. The plot also looks empty. with all the wasted space

```python
ax1 = sales_data.plot(legend = False)
ax1.set_xlabel('Month')
ax1.set_ylabel("Number of Sales")

ax1.set_ylim(150,1250)
plt.xticks([x for x in range(0,12)], sales_data.index, rotation=90)
plt.show()
```

### Rule 2: Hide what data you do show

```python
seed(123)
corr_data = pd.DataFrame([[x for x in range(1,20)],
                         [x * randint(1,6) for x in range(1,20)]]).transpose()

corr_data.columns = ['Number of Children', 'Soda per Day']

ax = corr_data.plot(style="o", ms = 2,
                            x = 'Number of Children', 
                            y = 'Soda per Day',
                            legend = False)
ax.set_facecolor('xkcd:grey')
plt.xticks([x for x in range(0,20)], [x for x in range(1, 21)])
plt.yticks([x * 20 for x in range(0,7)], [x * 20 for x in range(0, 7)])
plt.show()
```

Make things big enough and clear enough for your audience to see

```python
ax = corr_data.plot(style="o", ms = 7,
                            x = 'Number of Children', 
                            y = 'Soda per Day',
                    legend = False,
                   )

ax.set_xlabel('Number of Children', fontsize=18)
ax.set_ylabel('Soda per Day', fontsize=18)
plt.xticks([x for x in range(0,20)], [x for x in range(1, 21)])
plt.yticks([x * 20 for x in range(0,7)], [x * 20 for x in range(0, 7)])
plt.show()
```

### Rule 6: Emphasize the Trivial while Ignoring the Important

```python
pct_online_conversions = pd.DataFrame([['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'],
              [10, 12, 13, 15, 30, 35, 39, 38, 40, 38, 39, 41]]).transpose()

pct_online_conversions.columns = ['month', 'pct_conversions']

ax = pct_online_conversions.plot(legend = False,
                            title = 'Online Signups have Increased Each Month')

ax.set_xlabel('Month', fontsize=18)
ax.set_ylabel('Percent Online Signups (%)', fontsize=18)
plt.xticks([x * 2 for x in range(0,7)], [x * 2 for x in range(0, 7)])
plt.yticks([x * 5 for x in range(2,10)], [x * 5 for x in range(2, 10)])
plt.show()
```

### Rule 7: Alabama First (Order Matters)

```python
jp_sales_by_genre = vg_data[vg_data['Year'] == 2016].groupby('Genre').sum()['JP_Sales'].sort_values()
jp_sales_by_genre = jp_sales_by_genre.reset_index().rename(columns = {'JP_Sales':'JP_Sales_millions'})

jp_sales_by_genre.sort_values(by = 'Genre', inplace = True)

ax = jp_sales_by_genre.plot(kind = 'bar',
                       x = 'Genre',
                       y = 'JP_Sales_millions',
                       ylabel = 'Japan Sales in Millions',
                       xlabel = 'Genre',
                       legend = False)

ax.set_xlabel('Genre', fontsize=18)
ax.set_ylabel('Japan Sales (Millions)', fontsize=18)
plt.show()
```

```python
jp_sales_by_genre.sort_values(by = 'JP_Sales_millions', ascending = False, inplace = True)

ax = jp_sales_by_genre.plot(kind = 'bar',
                       x = 'Genre',
                       y = 'JP_Sales_millions',
                       ylabel = 'Japan Sales in Millions',
                       xlabel = 'Genre',
                       legend = False)

ax.set_xlabel('Genre', fontsize=18)
ax.set_ylabel('Japan Sales (Millions)', fontsize=18)

plt.show()
```

# A Few More Things

- Is your visualization representing the truth of the data?
- Are parts of your visualization taking away from the data?
- Is the scale of the data and the data range chosen correct?
- Are all axes labeled correctly and is other relevent information placed on the chart?
- Do the colors you chose help seperate or show the data is continuous? 
- Are the colors colorblind safe?
- Does the visualization you made tell the story that you want to portray?

## Representing the Truth of the Data

### Stacked Bar Chart

```python
percentage_sales = vg_data[(vg_data['Year'] == 2016) | (vg_data['Year'] == 2015)].groupby('Year').sum()[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]

global_2016 = percentage_sales.Global_Sales
percentage_sales['NA'] = percentage_sales.NA_Sales / global_2016
percentage_sales['EU'] = percentage_sales.EU_Sales / global_2016
percentage_sales['JP'] = percentage_sales.JP_Sales / global_2016
percentage_sales['Other'] = percentage_sales.Other_Sales / global_2016
percentage_sales.reset_index(inplace = True)
```

```python
percentage_sales
```

```python
# unpivot dataset
percentage_sales = pd.melt(percentage_sales, 
                           id_vars = ['Year'],
                           value_vars = ['NA', 
                                         'EU', 
                                         'JP', 
                                         'Other'])

percentage_sales.rename(columns = {'variable':'Sales',
                                   'value': 'Percent_Sales'}, inplace = True)
```

```python
percentage_sales
```

When you have more than 2 levels, stacked barcharts with percentages are not as useful since it is hard for the human eye to know where the baseline is.

```python
p9.ggplot(percentage_sales) \
 + p9.geom_bar(p9.aes(y = 'Percent_Sales', 
                      x = 'factor(Year)', 
                      fill = 'Sales'), stat="identity") \
 + p9.scale_y_continuous(labels=lambda l: ["%d%%" % (v * 100) for v in l]) \
 + p9.ylab('Percent of Global Sales') \
 + p9.xlab('Year')
```

Although this plot looks okay, imagine if the groups did not line up as nicely from  year to year. 

### Side by Side Bar Chart

A better way to compare categories year over year would be to use side by side bar charts

```python
plt = p9.ggplot(percentage_sales) \
     + p9.geom_bar(p9.aes(y = 'Percent_Sales', 
                      x = 'factor(Sales)',
                      group = 'factor(Year)',
                      fill = 'factor(Year)'), stat="identity", position = 'dodge') \
     + p9.scale_y_continuous(labels=lambda l: ["%d%%" % (v * 100) for v in l]) \
     + p9.guides(fill=p9.guide_legend(title="Year")) \
     + p9.scale_fill_manual(values = ['red', 'green']) \
     + p9.ylab('Percent of Global Sales') \
     + p9.xlab('Region')

plt.save('images/red_green.png')

plt

```

## Are the Colors we Use Good?

About 1 in 8 men are color blind in the United States. To make your plots easily readable for all audiences, it is important to think about what colors to use.
